```python
# Apply() method -- it takes fun as argument, and apply to all the values in the series or column
# df.apply(function,convert-dtype= True,args = (()))
import pandas as pd
d = {
     'name' : ['a','b','c','d','e'],
     'age': [56,70,40,52,76]
}

df = pd.DataFrame(d)
print(df)
# ceate a fun to check age
def age(n):
  if(n <60):
    return 'pass'
  return 'Fail'
df['pass-data'] = df['age'].apply(age,convert_dtype = True)
df





```

```python
# map()
# by using map method it transfroms the values in series(col) using specified mapping data function
# maps takes dictionary and series functions
# new_series = old_series.map(mapping-fun)
s = pd.Series(['A','B','C','D','E'])

grades = {'A':'Excellent','B':'Good','C':'Fair','D':'Pass','E':'Fail'}
mapped = s.map(grades)
print(mapped)

```

```python
df = pd.DataFrame({
                    'A':['Eve','Ellyse','Perry','Mic']*3,
                    'B':['Kelly','Darbie','Hanah',]*4,
                    'C':['Silvers','Becky','Ida','RJ','VK','DK']*2,
                    'D':np.random.randn(12),
                    'E':np.random.randn(12)
})
```

```python
# group by column "c", --"D",'E . sum()
df.groupby('C')[['D','E']].mean()
# syntax
# df.groupby('categorical column name')[[numerical column names]].aggregate functions
```

```python
# grouping by multiple columns label forms mutiindex
df.groupby(["A",'C']).sum()
```
